```python
import numpy, scipy
```

```python
a_array=numpy.linspace(0,4,100)
```

```python
f = lambda x,a: 1/numpy.sqrt(a**4-x**4)
```

```python
import scipy.integrate as sci
```

```python
T=[]
for a in a_array:
    g=lambda x: f(x,a)
    Ta=sci.quad(g,0,a)[0]
    T.append(2*numpy.sqrt(2)*Ta)
```

```python
from matplotlib import pyplot
```

```python
pyplot.plot(a_array,T)
```

```python
T
```

My physical intuition is that T shouldn't be constant, it should vary with the initial amplitude, which is the case shown 
here.

As a test, I ran the same algorithm but for the potential $V(x)=x^2$, to check if the period will be constant or not. 

```python
f2 = lambda x,a: 1/numpy.sqrt(a**2-x**2)
```

```python
T2=[]
for a in a_array:
    g=lambda x: f2(x,a)
    Ta=sci.quad(g,0,a)[0]
    T2.append(2*numpy.sqrt(2)*Ta)
```

```python
pyplot.plot(a_array,T2)
```

```python
T2
```

it can be seen that T is constant in the case $V(x)=X^2$
